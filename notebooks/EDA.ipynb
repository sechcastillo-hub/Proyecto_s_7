```python
# Cargar librerias
import pandas as pd
import numpy as np
import plotly.express as px

import plotly.io as pio 
pio.templates.default = "plotly"
```

```python
# Cargar dataset
df = pd.read_csv(r"C:\Users\GTX 1660Ti\Downloads\Proyecto_S_7\vehicles_us.csv")
df["type"] = df["type"].astype(str)
```

```python
# Seleccionar solo variables numéricas para poder calcular estadísticas
opciones = df.select_dtypes(include=np.number).columns.tolist()
```

```python
# Elegir 2 variables cualquiera (por ejemplo, las dos primeras: price y model_year)
v = (opciones[0], opciones[1])
```

```python
# Filtrar nulos en esas dos columnas para evitar errores de cálculo
df_clean = df.dropna(subset=list(v))

# Visualizar histograma de variable 1
hist_plot01 = px.histogram(
    df_clean, 
    x = v[0], 
    title = f"Distribución {v[0]}", 
    color = "type", 
    #marginal = "box", 
    width = 700, 
    height = 400
)
hist_plot01.show()
```

```python
# Calcular estadisticos de variable 1
prom1 = np.mean(df_clean[v[0]])
med1 = np.median(df_clean[v[0]])
desv1 = np.std(df_clean[v[0]])
print("Media: {:.1f}".format(prom1))
print("Mediana: {:.1f}".format(med1))
print("Desviación: {:.1f}".format(desv1))
```

```python
# Visualizar histograma de variable 2
hist_plot02 = px.histogram(
    df_clean, 
    x = v[1], 
    title = f"Distribución {v[1]}", 
    color = "type", 
    marginal = "box", 
    width = 700, 
    height = 400
)
hist_plot02.show()
```

```python
# Calcular estadisticos de variable 2
prom2 = np.mean(df_clean[v[1]])
med2 = np.median(df_clean[v[1]])
desv2 = np.std(df_clean[v[1]])
print("Media: {:.1f}".format(prom2))
print("Mediana: {:.1f}".format(med2))
print("Desviación: {:.1f}".format(desv2))
```

```python
# Visualizar dispersion entre variables
disp_plot = px.scatter(
    df_clean, 
    x = v[0], 
    y = v[1], 
    color = "type", 
    title = f"Dispersión {v[1]} vs. {v[0]}", 
    width = 700, 
    height = 400
)
disp_plot.show()
```

```python
# Calcular correlacion entre variables
correl = np.corrcoef(df_clean[v[0]], df_clean[v[1]])
print("Correlación de Pearson: {:.1%}".format(correl[0,1]))
```
